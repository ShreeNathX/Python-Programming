### Element-wise operations

```python
import numpy as np

a = np.array([1, 2, 3])
b = np.array([4, 5, 6])

print(a + b)
print(a - b)
print(a * b)
print(a / b)
```

### Scalar Operations

```python
import numpy as np

a = np.array([1, 2, 3])

print(a + 2)
print(a * 3)
```

### Comparison Operations

```python
import numpy as np

a = np.array([1, 2, 3, 4])

print(a > 2)
print(a == 2)
```

### Logical Operations

```python
import numpy as np

a = np.array([2, 3, 5, 7])

print((a > 2) & (a < 4))
print((a > 2) | (a < 4))
print(~(a < 4))
```

### Aggregation Operations

| Operation              | Function              |
|-----------------------|-----------------------|
| Sum                   | `np.sum(a)`           |
| Mean (Average)        | `np.mean(a)`          |
| Median                | `np.median(a)`        |
| Minimum               | `np.min(a)`           |
| Maximum               | `np.max(a)`           |
| Standard Deviation    | `np.std(a)`           |
| Variance              | `np.var(a)`           |
| Product               | `np.prod(a)`          |
| Cumulative Sum        | `np.cumsum(a)`        |
| Cumulative Product    | `np.cumprod(a)`       |
| Arg Min (index of min)| `np.argmin(a)`        |
| Arg Max (index of max)| `np.argmax(a)`        |
| Percentile            | `np.percentile(a, q)` |
| Quantile              | `np.quantile(a, q)`   |
| Count Non-zero        | `np.count_nonzero(a)` |
| Any (logical OR)      | `np.any(a)`           |
| All (logical AND)     | `np.all(a)`           |

```python
import numpy as np

a = np.array([2, 4, 6, 9])
b = np.array([5, 7, 3])

print(np.sum(a))
print(np.sum(a) + np.sum(b))
```

#### Axis Concept

```python
a = np.array([[1, 2],
             [3, 4]])

print(np.sum(a, axis = 0))       # Column-wise
print(np.sum(a, axis = 1))       # Row-wise
```

### Dot Product

```python
import numpy as np

a = np.array([1, 3, 4])
b = np.array([5, 2, 8])

print(np.dot(a, b))
```

### Matrix Multiplication

```python
import numpy as np

a = np.array([[2, 4],
             [5, 3]])

b = np.array([[3, 7],
             [4, 2]])

print(np.matmul(a, b))
print(a @ b)
```

### BroadCasting

```python
import numpy as np

a = np.array([[1, 2, 3],
             [4, 5, 6]])

b = np.array([1, 1, 2])         # NumPy automatically expands b

print(a + b)
```

### Indexed-based Operations

```python
import numpy as np

a = np.array([10, 30, 40, 20])

print(a[[1, 3]])
```

### Conditional Operations

```python
import numpy as np

a = np.array([1, 2, 4, 5])

print(a[a > 2])
```
